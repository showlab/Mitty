```python
import torch
from datasets.custom_dataset import CustomDataset
from torch.utils.data import DataLoader, Dataset


my_dataset = CustomDataset(
        video_root='/data/lxy/sqj/code/lang-segment-anything/my_videos',
        mask_root='/data/lxy/sqj/code/lang-segment-anything/my_masks',
        flow_root='/data/lxy/sqj/code/RAFT/my_flows',
        height=480,
        width=720,
        sample_n_frames=81,
    )
my_dataloader = DataLoader(my_dataset,
                            batch_size=2,
                            shuffle=False)

# for batch in my_dataset:
        
#     breakpoint()

for batch in my_dataloader:
    masked_flow = batch['masked_flow'] # B, 2, F, H, W
    # ---------------------------------------------------------------------------------------
    # 处理masked_flow 有问题！！！
    # breakpoint()
    mask_values = batch['mask_values'] # B, C, F, H, W
    area_s=mask_values[:, 0, :, :, :].unsqueeze(1).sum(dim=[-1, -2])
    refined_flow = masked_flow.sum(dim=(-1, -2)) / area_s  # # B, 2, F
    norm = torch.norm(refined_flow, dim=1, keepdim=True) + 1e-8  # [B, 1, N]
    # 单位方向 + 长度 # [B, 3, N] ---> [B, N, 3]
    refined_flow = torch.cat([refined_flow / norm, norm], dim=1).permute(0, 2, 1) # [B, N, 3]
    breakpoint()
```
